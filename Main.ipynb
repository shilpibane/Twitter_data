```python
import json
import pandas as pd
import numpy as np
```

```python
file=open('data.txt')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
PS = PorterStemmer()
```

```python
temp=[]
line=file.readline()
cnt=0
while line:
    if cnt%2==0:
        d1=json.loads(line)
        temp.append(d1['text'])
    line=file.readline()
    cnt+=1
```

```python
list1=[]
for text in temp:
    msg=text
    msg=re.sub('[^a-zA-Z]',' ',msg).lower().split()
    msg = ' '.join(msg)
    list1.append(msg)
```

```python
list1[67]
```

```python
list2=[]
for text in temp:
    msg=text
    msg=re.sub('[^a-zA-Z]',' ',msg).lower().split()
    msg = [PS.stem(word) for word in msg if not word in stopwords.words('english','Hindi')]
    msg = ' '.join(msg)
    list2.append(msg)
```

```python
list2[67]
```

```python
list2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features=1000)
X = CV.fit_transform(temp)
X=X.toarrayay()
```
